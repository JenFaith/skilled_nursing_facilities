```python
import numpy as np
import pandas as pd
import seaborn as sn
```

```python
df = pd.read_csv('data/NH_QualityMsr_MDS_Apr2022.csv', 
                 encoding='cp1252',
                 dtype={'Federal Provider Number': 'str'})
```

```python
df.info()
```

```python
# All Measure Types
print(df['Measure Description'].unique())
```

```python
#Separating out residents into two groups: Long vs Short-stay patients

long_stay = df[df['Measure Description'].str.contains('long-stay')]
short_stay = df[df['Measure Description'].str.contains('short-stay')]
```

```python
long_stay['Measure Description'].unique()
```

```python
short_stay[short_stay['Measure Description']=='Percentage of short-stay residents who newly received an antipsychotic medication'].sort_values(by = 'Four Quarter Average Score',
                                                                                                                                              ascending = False)
```

```python
long_stay[long_stay['Measure Description'].str.contains(
    'depressive')].sort_values(by = 'Q4 Measure Score')
```
